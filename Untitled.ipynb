## Data Cleaning & EDA

```python
df.isnull().sum()
```

```python
df = df[['service_area', 'subject_age', 'subject_race', 'subject_sex', 'type', 'arrest_made', 'outcome']]
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.rename(columns = {'subject_area':'area', 'subject_age':'age', 'subject_sex': 'sex', 'subject_race': 'race'}, inplace = True)
```

```python
# giving each race a value to test on
race = []
for row in df['race']:
        if row == 'white' : race.append(1)
        elif row == 'asian/pacific islander': race.append(2)
        elif row == 'hispanic': race.append(3)
        elif row == 'black':  race.append(4)
        elif row == 'other': race.append(5)
```

```python
# adding to dataframe
df['race_score'] = race
```

```python
# giving the different outcomes values based on how strong the outcome is.
outcome = []
for row in df['outcome']:
        if row == 'citaion' : outcome.append(1)
        elif row == 'warning': outcome.append(0)
        elif row == 'arrest': outcome.append(2)
        else:
            outcome.append(0)
```

```python
# adding to dataframe
df['outcome_score'] = outcome
```

```python
# giving the sex values of 1, 2
sex = []
for row in df['sex']:
        if row == 'male' : sex.append(1)
        elif row == 'female': sex.append(2)
        else:
            sex.append(0)
```

```python
#adding to df
df['sex_score'] = sex
```

## Modeling

```python
X = df[['race_score', 'outcome_score', 'sex_score', 'age']].copy()
sc = StandardScaler()
X_sc = sc.fit_transform(X)
```

### KMeans Clustering

```python
km = KMeans(n_clusters=3, random_state=42)
km.fit(X_sc)
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
km.predict(X_sc)
```

```python
df['cluster'] = km.labels_
```

```python
df.groupby('cluster').mean()
```

```python
centroids = pd.DataFrame(
    sc.inverse_transform(km.cluster_centers_),
    columns=["age", "race_score", "outcome_score", 'sex_score']
)
```

```python
plt.figure(figsize=(7,7));

colors = ["red", "green", "blue"]
df['color'] = df['cluster'].map(lambda p: colors[p])

ax = df.plot(    
    kind="scatter", 
    x="race_score", y="outcome_score",
    figsize=(10,8),
    c = df['color']
)

centroids.plot(
    kind="scatter", 
    x="race_score", y="outcome_score", 
    marker="*", c=["r", "g", "b"], s=550,
    ax=ax
);
```

```python
plt.figure(figsize=(7,7));

colors = ["red", "green", "blue"]
df['color'] = df['cluster'].map(lambda p: colors[p])

ax = df.plot(    
    kind="scatter", 
    x="sex_score", y="outcome_score",
    figsize=(10,8),
    c = df['color']
)

centroids.plot(
    kind="scatter", 
    x="sex_score", y="outcome_score", 
    marker="*", c=["r", "g", "b"], s=550,
    ax=ax
);
```

### RandomForrest

```python
features = ['age',
           'race_score',
           'sex_score']
X = df[features]
y = df['outcome_score']
```

```python
y.value_counts(normalize=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
np.random.seed(42)
n_estimators=100
rfc = RandomForestClassifier(n_estimators=n_estimators)
```

```python
etc = ExtraTreesClassifier(n_estimators=100)
```

```python
cv_scores = cross_val_score(rfc, X_train, y_train, cv=5)
print(cv_scores.mean())
```

```python
cv_scores = cross_val_score(rfc, X_test, y_test, cv=5)
print(cv_scores.mean())
```

```python
cv_scores = cross_val_score(etc, X_train, y_train, cv=5)
print(cv_scores.mean())
```

```python
cv_scores = cross_val_score(etc, X_test, y_test, cv=5)
print(cv_scores.mean())
```
